## Bipartite Graphs: 

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_nodes_from(['A', 'B', 'C'], partition=0)
G.add_nodes_from(['1', '2', '3'], partition=1)
G.add_edges_from([('A', '1'), ('A', '2'), ('A', '3'),
                   ('B', '1'), ('B', '2'), ('B', '3'),
                   ('C', '1'), ('C', '2'), ('C', '3')])

top = nx.bipartite.sets(G)[0]
pos = nx.bipartite_layout(G, top)
nx.draw(G, pos=pos, with_labels=True, node_color=['lightgreen', 'palegoldenrod', 'lightblue', 'orange', 'purple', 'pink'])
plt.show()
```

The graph shown is a bipartite graph with two partitions, each containing three nodes. The first partition includes nodes 'A', 'B', and 'C', while the second partition includes nodes '1', '2', and '3'. Each node in one partition is connected to every node in the other partition, forming a complete bipartite graph. Specifically, nodes 'A', 'B', and 'C' are connected to nodes '1', '2', and '3', respectively. The nodes are arranged using a bipartite layout, with nodes in the same partition aligned vertically. This graph is an example of a bipartite graph, illustrating how nodes can be divided into two distinct sets with edges only between nodes of different sets.

## Degree of a Vertex: 

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 0
# Use seed for reproducibility
G = nx.Graph()

# Add nodes
G.add_nodes_from([1,2,3,4,5,6,7])

edgelist = [(1,2),(1,3),(4,5), (2, 4), (1, 1), (4, 6), (1, 4), (7, 7)]
G.add_edges_from(edgelist)
pos = nx.spring_layout(G, seed=seed)  # Seed for reproducible layout
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

$deg(v_1)$ = 5

In this graph, vertex 1 is connected to vertices 2, 3, 1, and 4. There are four edges incident to vertex 1. Additionally, there is a self-loop on vertex 1, which contributes to its degree. Therefore, the degree of vertex 1 is 5.

## The Handshake Theorem: 

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a complete graph with 2 nodes
G = nx.complete_graph(4)

# Seed for reproducible layout
seed = 57
pos = nx.spring_layout(G, seed=seed)

# Draw the graph
nx.draw(G, pos=pos, with_labels=True)
plt.show()
```

$deg(v_0)$ = 3 $\\$
$deg(v_1)$ = 3 $\\$
$deg(v_2)$ = 3 $\\$
$deg(v_3)$ = 3 $\\$

Total Edges: 6
$deg(v_0)$ + $deg(v_1)$ + $deg(v_2)$ + $deg(v_3)$ = 12

Edges * 2 = Total Degree $\\$
6 * 2 = 12

Since the total degree of the graph is even, the graph exists.

## Connected Components: 
### Graph: 

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a connected graph
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1), (5, 2)])

# Plot the graph
plt.figure(figsize=(8, 6))
nx.draw(G, pos=nx.circular_layout(G), with_labels=True, node_size=500, node_color='pink', edge_color='black', width = 2)
plt.title("Connected Graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a connected graph
G = nx.Graph()
G.add_edges_from([(1, 2), (3, 4), (5, 1), (5, 2)])

component_path = [(1, 2), (5, 1), (5, 2)]

# Plot the graph
plt.figure(figsize=(8, 6))
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, node_color='pink')
nx.draw_networkx_edges(G, pos, edgelist = component_path, edge_color='black', width=2)
plt.title("Connected Component")
plt.show()
```

The connected component shown in the graph is a subgraph of the larger connected graph. It consists of three vertices: vertices 1, 2, and 5. These vertices are connected by the edges (1, 2), (5, 1), and (5, 2), forming a triangular arrangement within the graph.

A connected component is a maximal subset of vertices in a graph such that there exists a path between every pair of vertices within that subset. In this case, the vertices 1, 2, and 5 form a connected component because there is a path between every pair of these vertices. For example, vertex 1 is connected to vertex 2, vertex 2 is connected to vertex 5, and vertex 5 is connected back to vertex 1, forming a closed loop.

## Euler's Circuit: 

```python
import networkx as nx

G = nx.Graph()
G.add_edges_from([(0, 1), (1, 2), (2, 0), (0, 3), (3, 4), (4, 0)])

nx.draw_circular(G, with_labels=True)
print("Eulerian Circuit: 0-1-2-0-3-4-0")
```

An Eulerian circuit is a path in a graph that visits every edge exactly once and returns to the starting point. For a graph to have an Eulerian circuit, all of its vertices must have an even degree, except for at most two vertices that can have an odd degree.

## The Travelling Salesman Problem: 

```python
def tsp(graph, n):
    # Initialize the memoization table
    dp = [[float('inf') for _ in range(1 << n)] for _ in range(n)]

    # Base case: no cities to visit
    for i in range(n):
        dp[i][1 << i] = 0

    # Fill the memoization table
    for mask in range(1, 1 << n):
        for i in range(n):
            for j in range(n):
                if mask & (1 << j) and i != j:
                    dp[i][mask] = min(dp[i][mask], dp[j][mask ^ (1 << j)] + graph[i][j])

    # The minimum cost is the value at the bottom-right corner of the table
    return dp[0][(1 << n) - 1]

# Example usage:
graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]
n = len(graph)
print(tsp(graph, n))  # Output: 55
```

The TSP (Traveling Salesman Problem) is an optimization problem where the goal is to find the shortest possible route that visits every city exactly once and returns to the starting point.
