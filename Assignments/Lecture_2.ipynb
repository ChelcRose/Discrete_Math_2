To show that the expected value $E[Y] = n \cdot p_i$, we need to calculate the summation $\sum_{y} y \cdot P(y)$.

## Question 1
Recall that the joint probability distribution $P(y)$ is given by:

$$P(y) = \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot p_3^{y_3} \dots p_k^{y_k}$$

where $y = (y_1, y_2, \dots, y_k)$ and $y_1 + y_2 + \dots + y_k = n$.

Now, let's compute the expected value:

\begin{align} E[Y] &= \sum_{y} y \cdot P(y)
&= \sum_{(y_1, y_2, \dots, y_k)} (y_1 + y_2 + \dots + y_k) \cdot \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot p_3^{y_3} \dots p_k^{y_k}
&= \sum_{(y_1, y_2, \dots, y_k)} y_1 \cdot \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot p_3^{y_3} \dots p_k^{y_k} +
& \quad \sum_{(y_1, y_2, \dots, y_k)} y_2 \cdot \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot p_3^{y_3} \dots p_k^{y_k} +
& \quad \dots +
& \quad \sum_{(y_1, y_2, \dots, y_k)} y_k \cdot \frac{n!}{y_1!y_2!y_3!...y_k!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot p_3^{y_3} \dots p_k^{y_k}
&= n \cdot p_1 \cdot \sum_{(y_1, y_2, \dots, y_{k-1})} \frac{(n-1)!}{y_1!y_2!y_3!...y_{k-1}!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot p_3^{y_3} \dots p_{k-1}^{y_{k-1}} +
& \quad n \cdot p_2 \cdot \sum_{(y_1, y_2, \dots, y_{k-1})} \frac{(n-1)!}{y_1!y_2!y_3!...y_{k-1}!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot p_3^{y_3} \dots p_{k-1}^{y_{k-1}} +
& \quad \dots +
& \quad n \cdot p_k \cdot \sum_{(y_1, y_2, \dots, y_{k-1})} \frac{(n-1)!}{y_1!y_2!y_3!...y_{k-1}!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot p_3^{y_3} \dots p_{k-1}^{y_{k-1}}
&= n \cdot p_1 + n \cdot p_2 + \dots + n \cdot p_k
&= n \cdot (p_1 + p_2 + \dots + p_k)
&= n \cdot 1
&= n \end{align}

So, the expected value $E[Y] = n \cdot p_i$.

## Question 2:

If 500 adults are sampled randomly, the probability that the sample contains 100 persons between 18 and 24, 200 between 25 and 34, and 200 between the ages of 45 and 64 is given by the multinomial distribution:

$$P(Y_{18-24} = 100, Y_{25-34} = 200, Y_{45-64} = 200) = \frac{500!}{100! \cdot 200! \cdot 200!} \cdot (0.18)^{100} \cdot (0.23)^{200} \cdot (0.27)^{200}$$

To find the expected value for obtaining a person in the 65 and above age group, we use the formula $E[Y] = n \cdot p_i$:

$$E[Y_{65-100}] = 500 \cdot 0.16 = 80$$

So, the expected value for obtaining a person in the 65 and above age group is 80.

Question 3: Experience has shown that 30% of all persons afflicted by a certain illness recover. A drug company has developed a new medication. Ten people with the illness were selected at random and received the medication; nine recovered shortly thereafter. Suppose that the medication was absolutely worthless. What is the probability that at least nine of ten receiving the medication will recover?

Question 4: Suppose that a lot of 5000 electrical fuses contains 5% defectives. If a sample of 5 fuses is tested, find the probability of observing at least one defective.

## Question 1:
The probability distribution of the number of successes in $n$ trials is called the binomial distribution. The probability mass function (PMF) of the binomial distribution is given by:

$$P(Y=k) = \binom{n}{k} p^k q^{n-k}$$

where $n$ is the number of trials, $k$ is the number of successes, $p$ is the probability of success, and $q = (1-p)$ is the probability of failure. The term $\binom{n}{k}$ is the binomial coefficient, which counts the number of ways to choose $k$ successes from $n$ trials.

## Question 2:
To show that the expectation of the binomial distribution is $E[Y] = np$, we need to compute the summation $\sum_{k=0}^{n} k \cdot P(Y=k)$.

\begin{align} E[Y] &= \sum_{k=0}^{n} k \cdot P(Y=k)
&= \sum_{k=1}^{n} k \cdot \binom{n}{k} p^k q^{n-k}
&= \sum_{k=1}^{n} k \cdot \frac{n!}{k!(n-k)!} p^k q^{n-k}
&= np \cdot \sum_{k=1}^{n} \frac{(n-1)!}{(k-1)!(n-k)!} p^{k-1} q^{n-k}
&= np \cdot \sum_{j=0}^{n-1} \binom{n-1}{j} p^{j} q^{(n-1)-j}
&= np \cdot (p + q)^{n-1}
&= np \cdot 1
&= np \end{align}

So, the expectation of the binomial distribution is $E[Y] = np$.

## Question 3:
The probability that at least nine of ten receiving the medication will recover, assuming that the medication is absolutely worthless, is the same as the probability that nine or ten of the ten people recover by chance alone. This is a binomial distribution problem with $n=10$, $p=0.3$, and $k=9$ or $k=10$.

$$P(Y \geq 9) = P(Y=9) + P(Y=10) = \binom{10}{9} (0.3)^9 (0.7)^1 + \binom{10}{10} (0.3)^{10} (0.7)^0 \approx 0.0012 + 0.0000 \approx 0.0012$$

So, the probability that at least nine of ten receiving the medication will recover, assuming that the medication is absolutely worthless, is approximately 0.0012 or 0.12%.

## Question 4:
The probability of observing at least one defective in a sample of 5 fuses from a lot of 5000 electrical fuses containing 5% defectives is the complement of the probability of observing no defectives. This is a binomial distribution problem with $n=5$, $p=0.05$, and $k=0$.

$$P(Y \geq 1) = 1 - P(Y=0) = 1 - \binom{5}{0} (0.05)^0 (0.95)^5 \approx 1 - 0.7738 \approx 0.2262$$

So, the probability of observing at least one defective in a sample of 5 fuses is approximately 0.2262 or 22.62%.

Consider a probility distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

Question 1: Find a general formula for the expected value of this distribution.

Question 2: A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.

## Question 1:
The probability distribution given is the Poisson distribution with parameter $\lambda$. The expected value of the Poisson distribution is equal to its parameter, i.e., $E[Y] = \lambda$.

## Question 2:
The number of seedlings in ten 1-square-yard sampling regions is a Poisson distribution problem with $\lambda = 5$ and $n=10$. We want to find the probability that none of the regions will contain seedlings, i.e., $P(Y=0)$ where $Y$ is the number of seedlings in the ten sampling regions.

The number of seedlings in the ten sampling regions is the sum of the number of seedlings in each of the sampling regions, which are independent Poisson random variables. The sum of independent Poisson random variables is also a Poisson random variable with parameter equal to the sum of the parameters of the individual Poisson random variables.

Therefore, the number of seedlings in the ten sampling regions is a Poisson distribution with $\lambda = 5 \times 10 = 50$.
